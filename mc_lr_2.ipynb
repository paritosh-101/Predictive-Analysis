```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = 1
b = 3
n = 3000
m = 5000
```

```python
a3=[]
b3=[]
#np.random.seed(3)
#yold = 4 * np.random.randn(1,n)
x = 4 * np.random.randn(1,n)
```

```python
for i in range(m):
    
    e = np.random.randn(1,n)
    y = a + np.multiply(b,x) + e
    
    sig_x = np.sum(x)
    sig_y = np.sum(y)
    sig_x2 = np.sum(x * x)
    sig_xy = np.sum(x * y)
    
    j = ((sig_y * sig_x2) - (sig_x * sig_xy)) / (n*(sig_x2) - (np.square(sig_x)))
    k = (n*(sig_xy) - (sig_x * sig_y)) / (n*(sig_x2) - (np.square(sig_x)))
    
    a3.append(j)
    b3.append(k)
    
    x = x + np.random.randn()
```

```python
plt.hist(a3, density=True, bins=30, ec='white')
plt.title("alpha (a): sample size (n) = 3000, replications (m) = 5000")
plt.show()

print(round(np.median(a3),6), round(np.mean(a3),6), round(np.max(a3),6), round(np.min(a3),6), round(np.std(a3),6))
```

```python
plt.hist(b3, density=True, bins=30, ec='white')
plt.title("beta (b): sample size (n) = 3000, replications (m) = 5000")
plt.show()

print(round(np.median(b3),6), round(np.mean(b3),6), round(np.max(b3),6), round(np.min(b3),6), round(np.std(b3),6))
```

```python
a = 1
b = 3
n = 1000
m = 5000
p = 3
a4=[]
b4=[]
np.random.seed(4)
x = 4 * np.random.randn(1,n)
```

```python
for i in range(m):
    
    e = np.random.randn(1,n)
    y = a + np.multiply(b,np.power(x,p)) + e
    
    sig_x = np.sum(x)
    sig_y = np.sum(y)
    sig_x2 = np.sum(x * x)
    sig_xy = np.sum(x * y)
    
    j = ((sig_y * sig_x2) - (sig_x * sig_xy)) / (n*(sig_x2) - (np.square(sig_x)))
    k = (n*(sig_xy) - (sig_x * sig_y)) / (n*(sig_x2) - (np.square(sig_x)))
    
    a4.append(j)
    b4.append(k)
```

```python
plt.hist(a4, density=True, bins=30, ec='white')
plt.title("alpha (a): sample size (n) = 1000, replications (m) = 5000")
plt.show()

print(round(np.median(a4),6), round(np.mean(a4),6), round(np.max(a4),6), round(np.min(a4),6), round(np.std(a4),6))
```
